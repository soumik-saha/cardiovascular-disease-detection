```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from imblearn.combine import SMOTEENN

# Separate the features and target variable
X_train = df_selected.drop("cardio", axis=1)  # Use 'cardio' as the target
y_train = df_selected["cardio"]

# Perform SMOTEENN oversampling on the resampled data
smote_enn = SMOTEENN(random_state=42)
X_train_resampled_smoteenn, y_train_resampled_smoteenn = smote_enn.fit_resample(X_train, y_train)

# Perform RandomOverSampler on the training data
random_oversampler = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = random_oversampler.fit_resample(X_train_resampled_smoteenn, y_train_resampled_smoteenn)

# Combine the resampled features and target variable into a single DataFrame
resampled_data = pd.concat([X_train_resampled, y_train_resampled], axis=1)

# Separate the instances with 0 and 1 values for the target variable 'cardio'
data_0 = resampled_data[resampled_data["cardio"] == 0]
data_1 = resampled_data[resampled_data["cardio"] == 1]

# Sample 3000 instances from each class
data_0_sample = data_0.sample(n=3000, random_state=42)
data_1_sample = data_1.sample(n=3000, random_state=42)

# Concatenate the sampled instances into a single DataFrame
first_csv_data = pd.concat([data_0_sample, data_1_sample])

# Drop the sampled instances from the original data to get the remaining instances
remaining_data_0 = data_0.drop(data_0_sample.index)
remaining_data_1 = data_1.drop(data_1_sample.index)
second_csv_data = pd.concat([remaining_data_0, remaining_data_1])

# Save the DataFrames to CSV files
first_csv_data.to_csv("test.csv", index=False)
second_csv_data.to_csv("heart_disease_train_resampled.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Load the training dataset
df_train = pd.read_csv("heart_disease_train_resampled.csv")

# Separate the features and target variable
X = df_train.drop("cardio", axis=1)
y = df_train["cardio"]

# Normalize the data
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Define the model (choose your preferred model, e.g., DNN, CNN, or RNN)
# Here's an example with a DNN:
def create_nn_model(input_shape):
    model_dnn = Sequential()
    model_dnn.add(Dense(64, activation='relu', input_shape=(input_shape,)))
    model_dnn.add(Dense(32, activation='relu'))
    model_dnn.add(Dense(1, activation='sigmoid'))
    return model_dnn

nn_model = create_nn_model(X_normalized.shape[1])
nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Implement Early Stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model
history = nn_model.fit(
    X_train,
    y_train,
    epochs=50,  # Adjust the number of epochs as needed
    batch_size=32,
    validation_data=(X_val, y_val),
    callbacks=[early_stopping]
)

# Analyze the training history to detect underfitting or overfitting
plt.figure(figsize=(12, 4))

# Plot training & validation accuracy values
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam  # Fix the optimizer import
import keras_tuner as kt

# Load the training dataset
df_train = pd.read_csv("heart_disease_train_resampled.csv")

# Separate the features and target variable
X = df_train.drop("cardio", axis=1)
y = df_train["cardio"]

# Normalize the data
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Hyperparameter tuning function
def build_model(hp):
    model = Sequential()

    # Add an Input layer to replace input_shape in Dense layers
    model.add(Input(shape=(X_train.shape[1],)))

    # Tuning the number of units in the first Dense layer
    hp_units1 = hp.Int('units1', min_value=32, max_value=512, step=32)
    model.add(Dense(units=hp_units1, activation='relu'))

    # Tuning the number of layers and units in each layer
    for i in range(hp.Int('num_layers', 1, 3)):  # Allows between 1 and 3 hidden layers
        hp_units = hp.Int(f'units_{i+2}', min_value=32, max_value=512, step=32)
        model.add(Dense(units=hp_units, activation='relu'))

    # Output layer
    model.add(Dense(1, activation='sigmoid'))

    # Tuning optimizer and learning rate
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    model.compile(optimizer=Adam(learning_rate=hp_learning_rate),  # Correct Adam import
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    return model

# Initialize the tuner
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,  # You can increase this for more exhaustive search
    executions_per_trial=2,
    directory='keras_tuner_dir',
    project_name='heart_disease_dnn')

# Early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Perform the hyperparameter search
tuner.search(X_train, y_train, epochs=50, validation_data=(X_val, y_val), callbacks=[early_stopping])

# Get the optimal hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

# Build the best model
best_model = tuner.hypermodel.build(best_hps)

# Train the model with the optimal hyperparameters
history = best_model.fit(X_train, y_train, epochs=50, validation_data=(X_val, y_val), callbacks=[early_stopping])

# Plotting the training history to visualize performance improvements
plt.figure(figsize=(12, 4))

# Plot training & validation accuracy values
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.show()

# Evaluate the model
val_loss, val_accuracy = best_model.evaluate(X_val, y_val)
print(f"Best validation accuracy: {val_accuracy:.4f}")
```
